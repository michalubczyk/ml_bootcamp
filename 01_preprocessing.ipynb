```python
import pandas as pd
import numpy as np
import sklearn
sklearn.__version__
```

### Generowanie danych

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

### Tworzenie kopii danych

```python
df = df_raw.copy()
df.info()
```

### Zmiana typu danych

```python
for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
df.describe()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

```python
df
```

### LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])
```

```python
le.fit_transform(df['bought']) # powyzszy zapis w krotszej formie
```

```python
le.classes_ # postac mapowania
```

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

```python
le.inverse_transform(df['bought']) # operacja odwrotna do le.fit_transform
```

### OneHotEncoder (zaleca się stosowanie get_dummies z biblioteki Pandas)

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']])
```

```python
encoder.categories_
```

```python
encoder = OneHotEncoder(drop='first', sparse_output=False) # n-1 kolumn po kodowaniu
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

```python
df
```

### Pandas get_dummies() (alternatywa dla OneHotEncoder)

```python
pd.get_dummies(data=df, drop_first=True) # zmienne 'category' zostaly zakodowane
```

### Standaryzacja
std() - pandas nieobciazony<br>
std() - numpy obciazony

```python
print(f"{df['price']}\n")
print(f"Srednia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}")
```

```python
(df['price'] - df['price'].mean()) / df['price'].std()
```

```python
from sklearn.preprocessing import scale
scale(df['price']) # alternatywna standaryzacja (wartosci roznia sie od powyzszych)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']]) # zalecana alternatywa
```

```python
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```
